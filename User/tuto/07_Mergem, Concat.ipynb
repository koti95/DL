```python
df5 = df.T.iloc[:7,:3]
df6 = df.T.iloc[6:,2:5]

display(df5)
display(df6)
```

```python
df5 = df5.rename(columns=df5.iloc[0])
df5 = df5.drop(df5.index[0])
df5

df6 = df6.rename(columns=df6.iloc[0])
df6 = df6.drop(df6.index[0])
df6
```

```python
Ans = pd.merge(df5,df6,on='Algeria',how='inner')
Ans
```
