```python
import pandas as pd
import re
from konlpy.tag import Okt
from collections import Counter
import pytagcloud

okt = Okt()
```

```python
data = pd.read_csv('유라_중복제거.csv', encoding='cp949')
data
```

```python
data1 = data.iloc[:,1]
data1
```

```python
noun = okt.nouns(data1[2])
noun
```

```python
noun = []
for i in range(len(data1)):
    noun.append(okt.nouns(data1[i]))
noun
```

```python
noun = sum(noun, [])
noun
```

```python
count = Counter(noun)

noun_list = count.most_common(100)
for i in noun_list:
    print(i)
```
